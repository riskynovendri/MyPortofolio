RFM Calculation

```
#Func to get Recencies
def getR(df) :
  Recencies=[]
  for idx, row in df.iterrows() :
    list_col = df.columns
    for col in reversed(list_col):
      if row[col]==1:
        Recencies.append(col)
        break
  return Recencies
```

```
R = getR(Day) #Get R Value
```

```
F = Day.sum(axis=1) #Get Frequent Value
```

```
Day['R'] = R
Day['F'] = F
```
