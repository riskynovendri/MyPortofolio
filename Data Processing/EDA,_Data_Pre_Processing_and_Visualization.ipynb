```python
min(df['qty']) #cek minimum value
```

```python
df_no_min = df[df['qty'] >= 0] #filter data tanpa minus pada qty
```

```python
min(df_no_min['qty'])
```

```python
df_no_min['total_harga'] = df_no_min['qty'] * df_no_min['harga_awal'] #membuat kolom baru total harga
```

```python
df_no_min[df['qty']>1] #cek data 
```

```python
df_no_min #cek hasil akhir
```

```python
df_sku = df_no_min[['nama_sku','qty','total_harga']].groupby(by='nama_sku').sum() #aggregasi data
```

```python
df_sku
```

```python
df_sku_sorted = df_sku.sort_values(by='qty',axis=0,ascending=False)
```

```python
df_sku_sorted.head(10) #cek top 10 produk
```

```python
#Cek Jumlah perbanding sumber
plt.figure(figsize=[10,10])
sns.countplot(df_no_min['sumber']).set_title('Performa Sumber')
```

```python
#visualisasi Top 10 Product
plt.figure(figsize=[10,10])
sns.barplot(x='qty',y=df_sku_sorted.head(10).index,data=df_sku_sorted.head(10)).set_title('TOP 10 SKU Paling Banyak Terjual')
```

```python
df_sumber = df_no_min.groupby(by=['nama_sku','sumber'], as_index=False).sum().sort_values(by='qty',axis=0,ascending=False) #Aggregasi data
```
