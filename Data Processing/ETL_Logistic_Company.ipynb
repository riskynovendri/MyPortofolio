## Data Merge -> Menggabungkan ketiga data kota

## Membuat kolom factory

```python
Df_merge["SL AHM No"].str[:4] # Mengambil nilai data prabrik dari SL AHM No
```

```python
#Fungsi untuk mengecek datanya
def locate_factory(data):
  factory=''
  if data == '1100':
    factory = "Pabrik Sunter"
  elif data == '1200':
    factory="Pabrik Pegangsaan"
  elif data == '1300' or '1350':
    factory = "Pabrik Cibitung"
  elif data == '1700' or '1800':
    factory = "Pabrik Dawuan"
  return factory
```

```python
# Membuat kolom factory dari nilai SL AHM No
Df_merge['Factory'] = Df_merge["SL AHM No"].str[:4].apply(lambda x: locate_factory(x))
```

```python
Df_merge
```

# Snowflake Schema implementation

```python
# Merename kolom sesuai shema
Df_merge.rename(columns={"Engine No":"engine_no","NO. CONTAINER":"no_container","Dooring Date":"dooring_date","NAMA TYPE":"unit_name","WARNA":"unit_color","KODE 3 DIGIT":"unit_material",
                         "SL TMS NO":"sl_tms_no","Aging Unit di Pool":"aging_pool","Aging Unit Intransit":"aging_intransit","SL AHM No":"sl_ahm_no","SL AHM Date":"sl_ahm_date",
                         "Lokasi":"status","Stuffing Date":"stuffing_date"},inplace=True)
```

```python
# Merename kolom sesuai shema (2)
Df_merge.rename(columns={"Kapal":"ship_name","Truck":"truck_name","Factory":"factory_name","Destination":"destination_name"},inplace=True)
```

```python
import uuid # Untuk membuat libry generate ID
import string 
import random
# Membuat Fungsi gerate ID versi lainnya
def id_generator(size=6, chars=string.ascii_uppercase + string.digits):
    return ''.join(random.SystemRandom().choice(chars) for _ in range(size))
```
