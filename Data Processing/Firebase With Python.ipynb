## Import file csv

```python
#Read data csv menggunakan Spark
path_spark_jateng = COLLECTION_PATH+"/jateng.csv"
df_spark_jateng = spark.read.csv(path_spark_jateng, sep=",", header=True)
df_spark_jateng = df_spark_jateng.drop('_c0')
df_spark_jateng.show()
```

```python
#Melakukan seleksi data yang diperlukan
df_spark_jateng = df_spark_jateng.select(df_spark_jateng["Date"], df_spark_jateng["Location"], df_spark_jateng["New Cases"], df_spark_jateng["New Deaths"], \
                                       df_spark_jateng["New Recovered"], df_spark_jateng["Total Cases"], df_spark_jateng["Total Deaths"], df_spark_jateng["Total Recovered"])
df_spark_jateng.show()
```

## Import file excel

```python
#Read file excel menggunakan spark
path_spark_jatim = COLLECTION_PATH + "/jatim.xlsx"
df = pd.read_excel(path_spark_jatim)
df_spark_jatim = spark.createDataFrame(df)
df_spark_jatim = df_spark_jatim.drop('Unnamed: 0')
df_spark_jatim.show()
```

```python
#Seleksi data
df_spark_jatim = df_spark_jatim.select(df_spark_jatim["Date"], df_spark_jatim["Location"], df_spark_jatim["New Cases"], df_spark_jatim["New Deaths"], \
                                       df_spark_jatim["New Recovered"], df_spark_jatim["Total Cases"], df_spark_jatim["Total Deaths"], df_spark_jatim["Total Recovered"])
df_spark_jatim.show()
```

## Import file json

```python
#Read data format json menggunakan Spark
path_spark_jabar = COLLECTION_PATH + "/jabar.json"
df_jabar = pd.read_json(path_spark_jabar)
df_spark_jabar = spark.createDataFrame(df_jabar)
df_spark_jabar.show()
```

```python
#Seleksi data
df_spark_jabar = df_spark_jabar.select(df_spark_jabar["Date"], df_spark_jabar["Location"], df_spark_jabar["New Cases"], df_spark_jabar["New Deaths"], \
                                       df_spark_jabar["New Recovered"], df_spark_jabar["Total Cases"], df_spark_jabar["Total Deaths"], df_spark_jabar["Total Recovered"])
df_spark_jabar.show()
```

## Import file xml

```python
#Read data XML
path_data_jakarta = COLLECTION_PATH+"/jakarta(1).xml"
df_jakarta = pd.read_xml(path_data_jakarta)
df_spark_jakarta = spark.createDataFrame(df_jakarta)
df_spark_jakarta.show()
```

```python
#Seleksi Data
df_spark_jakarta = df_spark_jakarta.select(df_spark_jakarta["Date"], df_spark_jakarta["Location"], df_spark_jakarta["New_Cases"], df_spark_jakarta["New_Deaths"], \
                                       df_spark_jakarta["New_Recovered"], df_spark_jakarta["Total_Cases"], df_spark_jakarta["Total_Deaths"], df_spark_jakarta["Total_Recovered"])
df_spark_jakarta.show()
```

## Merge DataFrame

```python
#Melihat jumlah data masing-masing data
print(df_spark_jateng.count())
print(df_spark_jatim.count())
print(df_spark_jabar.count())
print(df_spark_jakarta.count())
```

```python
#Menggabungkan dataset yang berbeda menjadi satu
df_merge = df_spark_jateng.union(df_spark_jatim)
df_merge = df_merge.union(df_spark_jabar)
df_merge = df_merge.union(df_spark_jakarta)
df_merge.show()
```

## Export to Firebase

```python
#Melakukan format database
fb_app.delete('/', '')
print('Record Deleted')
```

```python
#memasukkan data kedalam firebase
dataCollect = df_merge.collect()

for row in dataCollect: #Iterasi data kedalam format json object
    data = {
        'Date': row['Date'], 
        'Location': row['Location'],
        'New Cases':  row['New Cases'],
        'New Deaths':  row['New Deaths'], 
        'New Recovered':  row['New Recovered'], 
        'Total Cases':  row['Total Cases'], 
        'Total Deaths':  row['Total Deaths'], 
        'Total Recovered':  row['Total Recovered']
      }
    result = fb_app.post('/', data) #Post data setiap iterasi dari perulangan sebelumnya
```
