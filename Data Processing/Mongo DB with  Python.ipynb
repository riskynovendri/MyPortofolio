```python
#CSV to JSON Conversion
csvfile = open('/content/drive/MyDrive/Remedial Big Data/estonia-passenger-list.csv', 'r')
reader = csv.DictReader( csvfile )
header= [ "PassengerId", "Country", "Firstname", "Lastname","Sex","Age","Category","Survived"]
Collection = db["listpenumpang"]
for each in reader: #Iterasi Baris pada CSV kedalam format JSON
    row={}
    for field in header:
        row[field]=each[field]

    Collection.insert_one(row) #Import Baris Ke MongoDB
```
